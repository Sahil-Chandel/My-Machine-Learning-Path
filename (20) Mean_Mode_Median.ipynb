Measure of Central Tendency
 1. Mean
 2. Median
 3. Mode

```python
ages = [23,24,32,45,12,43,67,45,32,56,32,100]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
```

```python
import seaborn as sns
sns.boxplot(ages)
```

## 5 Number Summary

```python
import numpy as np
q1,q3=np.percentile(ages,[25,75])
```

```python
print(q1)
print(q3)
```

```python
# To check outliers [Lower Fence - Higher Fence]
IQR=q3-q1
print(IQR)
lower_fence = q1-1.5*(IQR)
higher_fence = q3 +1.5*(IQR)
print(lower_fence, higher_fence)
```

## Measure of dispersion
 1. varience
 2. Standard Deviation

```python
statistics.variance(ages)
```

```python
np.var(ages,axis=0)
```

```python
def varience(data):
    n=len(ages)
    ## mean of the data
    mean=sum(data)/n
    ## varience
    deviation=[(x - mean) ** 2 for x in data]
    varience=sum(deviation)/n-1
    return varience
```

```python
varience(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

# Check whether Distribution is Normal Distribution


```python
## If you want to check whether feature is gaussian or normal distributed
## Q-Q Plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample, kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## Create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
plot_data_log(sample)
```
